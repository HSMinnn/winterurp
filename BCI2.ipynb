```python
#prob.4

num_files = 32  # 총 파일 개수
num_rows = 40
all_rows = [[] for _ in range(num_rows)]


for i in range(1, 33):
    mat_file_name = "./DEAP/s{:02d}.mat".format(i)
    mat_file = scipy.io.loadmat(mat_file_name)
    print(f'{mat_file_name} has been loaded.')
    
    numpy_file = np.array(mat_file['labels'])
    
    for i in range(40):
        first_element = numpy_file[i][0]
        all_rows[i].append(first_element)


for row_index, row_data in enumerate(all_rows):
    plt.figure()
    plt.hist(row_data, bins=10, color='blue', alpha=0.7)  # 원하는 스타일 적용
    plt.xlabel("valence")
    plt.ylabel("Frequency")

plt.show()
```

```python
#prob.5

for file_index in range(1, 33):  
    mat_file_name = "./DEAP/s{:02d}.mat".format(file_index)
    mat_file = scipy.io.loadmat(mat_file_name)
    numpy_file = np.array(mat_file['labels'])

    Valence_list = []
    Arousal_list = []
    
    for row_index in range (40) :
        result = round(numpy_file[row_index][0])
        Valence_list.append(result)

    
    plt.figure()
    plt.hist(Valence_list, bins=10, color='blue', alpha=0.7)  # 원하는 스타일 적용
    plt.xlabel("Rounded Valence")
    plt.ylabel("Frequency")


    for row_index in range(40) :
        result = round(numpy_file[row_index][1])
        Arousal_list.append(result)

    plt.figure()
    plt.hist(Arousal_list, bins=10, color='blue', alpha=0.7)  # 원하는 스타일 적용
    plt.xlabel("Rounded Arousal")
    plt.ylabel("Frequency")
    
plt.show()

    
    
```
